```python
# -*- coding: utf-8 -*-
import matplotlib.pyplot as plt
import copy
from sklearn import cluster
from sklearn import datasets

# iris データをロード
iris = datasets.load_iris()
data1 = iris["data"]
data2 = copy.deepcopy(iris["data"])

# 学習 → クラスタの生成
model = cluster.AgglomerativeClustering(n_clusters=3, linkage="ward")
model.fit(data1)

# 学習結果のラベル取得
labels = model.labels_

### グラフの描画

# データ定義
x_index = 2
y_index = 3

data_x1 = data1[:, x_index]
data_y1 = data1[:, y_index]

x_label = iris["feature_names"][x_index]
y_label = iris["feature_names"][y_index]

plt.figure(figsize=(20, 10))

# 散布図描画
plt.subplot(1, 2, 1)
plt.scatter(data_x1[labels == 0], data_y1[labels == 0], c='black', alpha=0.3, s=100, marker="o")
plt.scatter(data_x1[labels == 1], data_y1[labels == 1], c='black', alpha=0.3, s=100, marker="^")
plt.scatter(data_x1[labels == 2], data_y1[labels == 2], c='black', alpha=0.3, s=100, marker="*")

# 軸ラベルとタイトルの設定
plt.xlabel(x_label, fontsize='xx-large')
plt.ylabel(y_label, fontsize='xx-large')
plt.title("AgglomerativeClustering(ward)", fontsize='xx-large')

model2 = cluster.KMeans(n_clusters=3).fit(data2)

# 学習結果のラベル取得
labels2 = model.labels_

# グラフの描画
# データ定義
x_index = 2
y_index = 3

data_x2 = data2[:, x_index]
data_y2 = data2[:, y_index]

x_label = iris["feature_names"][x_index]
y_label = iris["feature_names"][y_index]


# 散布図描画
plt.subplot(1, 2, 2)
plt.scatter(data_x2[labels == 0], data_y2[labels == 0], c='black', alpha=0.3, s=100, marker="o")
plt.scatter(data_x2[labels == 1], data_y2[labels == 1], c='black', alpha=0.3, s=100, marker="^")
plt.scatter(data_x2[labels == 2], data_y2[labels == 2], c='black', alpha=0.3, s=100, marker="*")

# 軸ラベルとタイトルの設定
plt.xlabel(x_label, fontsize='xx-large')
plt.ylabel(y_label, fontsize='xx-large')
plt.title("k-means", fontsize='xx-large')

plt.show()
```
